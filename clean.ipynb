```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("fear_greed_index.csv")
df.head()
```

```python
print(df.shape)
print(df.info())
print(df.describe())
```

```python
print("The null values are:\n",df.isnull().sum())
print()
print(df.isna().sum())
print()
print("The duplicate values are:\n",df.duplicated().sum())
```

```python
Q1 = df['value'].quantile(0.25)
Q3 = df['value'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

print(f"Value range: {lower} to {upper}")
print(f"Outliers: {((df['value'] < lower) | (df['value'] > upper)).sum()}")
```

```python
df["timestamp"] = pd.to_datetime(df["timestamp"],unit='s').dt.normalize()
```

```python
df.to_csv("cleaned_fear_greed.csv", index=False)
print("Saved!")
```
