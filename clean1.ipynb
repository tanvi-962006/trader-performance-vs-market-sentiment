```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
df = pd.read_csv("historical_data.csv")
df.head()
```

```python
print(df.shape)
print(df.info())
print(df.describe())
```

```python
print(df.isnull().sum())
print()
print(df.isna().sum())
print()
print("The sum of duplicated values are:\n",df.duplicated().sum())
```

```python


for col in ['Execution Price', 'Size USD', 'Closed PnL', 'Fee']:
    z_scores = np.abs(stats.zscore(df[col]))
    outliers = (z_scores > 3).sum()
    print(f"{col}: {outliers} outliers ({(outliers/len(df))*100:.1f}%)")
```

```python

for col in ['Execution Price', 'Size USD', 'Closed PnL', 'Fee']:
    lower = df[col].quantile(0.01)
    upper = df[col].quantile(0.99)
    df[col] = df[col].clip(lower, upper)

print("Outliers capped successfully!")
```

```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'], unit='ms').dt.normalize()
df['Timestamp IST'] = pd.to_datetime(df['Timestamp IST'], format='%d-%m-%Y %H:%M').dt.normalize()
```
